```python
from transformers import pipeline

# Load zero-shot classification pipeline
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Define material descriptions
descriptions = ["lightweight, corrosion-resistant, used in aerospace", "heat resistant, high tensile strength, used in construction"]

# Define candidate labels
labels = ["aluminum", "titanium", "steel", "carbon fiber", "concrete"]

# Predict material properties
results = [classifier(description, candidate_labels=labels) for description in descriptions]
for res in results:
    print(res)
```
